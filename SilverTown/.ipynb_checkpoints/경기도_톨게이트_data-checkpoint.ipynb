```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./data/톨게이트.csv',
                   usecols={'영업소명', '주소'},
                   encoding='cp949',
                   low_memory = False)
```

```python
data
```

```python
res = data.주소.str.contains('경기', na=False)
data_G = data[res]
data_G
```

```python
addr = data_G.주소.str.slice(start=0, stop=8)
addr
addr.value_counts()

data_G.주소 = addr
```

```python
data_G.주소.value_counts()
```

```python
data_cnt = pd.DataFrame(data_G.주소.value_counts())
```

```python
data_cnt.rename(columns={'주소': '톨게이트'}, inplace = True)
```

```python
data_cnt.to_csv('./data/경기도_톨게이트_data')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('./data/지하철.csv',
                   usecols = ['SBW_STATN_NM', 'ROAD_NM_ADDR'],
                  low_memory=False)
```

```python
data
```

```python
res = data.ROAD_NM_ADDR.str.contains('경기', na=False)
data_subway = data[res]
data_subway

```

```python
data_subway.to_csv('./data/수도권(지하철).csv')
```

```python
data_G = data_subway.rename(columns={'ROAD_NM_ADDR':'주소'})
```

```python
addr = data_G.주소
addr
```

```python
addr = addr.str.slice(start=0, stop=8)
addr.value_counts()
```

```python
subway_G = pd.DataFrame(addr.value_counts())
```

```python
subway_G
```

```python
subway_G.rename(columns={'count':'지하철 역'},inplace=True)
```

```python
subway_G.index.name = '지역'
```

```python
subway_G
```

```python
subway_G.to_csv('./data/경기도_지하철역_data.csv')
```

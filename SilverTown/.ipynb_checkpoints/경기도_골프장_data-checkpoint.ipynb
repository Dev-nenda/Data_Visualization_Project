```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('./data/업종별골프장현황.csv', 
                  encoding='cp949',
                  low_memory=False)
```

```python
data.head()
```

```python
data.drop(columns={'정제우편번호', '정제WGS84경도', '정제WGS84위도', '정제도로명주소', '골프장면적', '골프장홀수'}, inplace=True)
```

```python
data
```

```python
res = data.정제지번주소.str.contains('경기', na=False)
data_G = data[res]
data_G
```

```python
addr = data_G.정제지번주소.str.slice(start=0, stop=8)
addr
addr.value_counts()

data_G.정제지번주소 = addr
```

```python
data_G
```

```python
data_G.정제지번주소.value_counts()
```

```python
data_G_cnt = pd.DataFrame(data_G.정제지번주소.value_counts())
```

```python
data_G_cnt
```

```python
data_G_cnt.index.name = '주소'
```

```python
data_G_cnt.rename(columns={'정제지번주소' : '골프장'}, inplace=True)
```

```python
data_G_cnt
```

```python
data_G_cnt['골프장'] = 1
        
```

```python
data_G_cnt
```

```python
data_G_cnt.to_csv('./data/경기도_골프장_유무_data.csv')
```

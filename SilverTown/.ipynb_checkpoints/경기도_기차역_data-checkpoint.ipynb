```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_station = pd.read_csv('./data/기차역.csv')
```

```python
train_station
```

```python
res = train_station.주소.str.contains('경기', na=False)
train_G = train_station[res]
train_G
```

```python
addr = train_G.주소.str.slice(start=0, stop=8)
addr
train_G.주소 = addr
```

```python
train_G.drop(columns={'Unnamed: 0'}, inplace = True)
```

```python
train_G
```

```python
train_G.주소.value_counts()
```

```python
train_G = pd.DataFrame(train_G.주소.value_counts())
```

```python
train_G
```

```python
train_G.rename(columns={'주소':'기차역'}, inplace=True)
```

```python
train_G
```

```python
train_G.to_csv('./data/경기도_기차역_data.csv')
```

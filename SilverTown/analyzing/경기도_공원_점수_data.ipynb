```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('./data/공원목록_수정.csv', 
                  low_memory=False)
```

```python
data
```

```python
res = data.소재지지번주소.str.contains('경기', na=False)
data_G = data[res]
```

```python
data_G
```

```python
data_G = data_G.drop(columns={'Unnamed: 0', '공원구분', '관리기관명'})
```

```python
data_G
```

```python
data_G = data_G.fillna(0)
```

```python
data_G = data_G.drop(columns={'공원보유시설(유희시설)', '공원보유시설(기타시설)'})
```

```python
data_G
```

```python
tf = data_G['공원보유시설(운동시설)'] != 0
```

```python
tf
```

```python
data_G['공원보유시설(운동시설)']=tf
```

```python
tf = data_G['공원보유시설(운동시설)'].replace(True, 1).replace(False, 0)
```

```python
tf
```

```python
data_G['공원보유시설(운동시설)']=tf
```

```python
data_G
```

```python
tf = data_G['공원보유시설(편익시설)'] != 0
```

```python
data_G['공원보유시설(편익시설)'] = tf
```

```python
data_G['공원보유시설(편익시설)'] = data_G['공원보유시설(편익시설)'].replace(True, 1).replace(False, 0)
```

```python
data_G
```

```python
tf = data_G['공원보유시설(교양시설)'] != 0
```

```python
data_G['공원보유시설(교양시설)'] = tf
```

```python
data_G['공원보유시설(교양시설)'] = data_G['공원보유시설(교양시설)'].replace(True, 1).replace(False, 0)
```

```python
data_G
```

```python
data_G['종합점수'] = data_G['공원보유시설(운동시설)'] + data_G['공원보유시설(편익시설)'] + data_G['공원보유시설(교양시설)']
```

```python
data_G
```

```python
addr = data_G.소재지지번주소.str.slice(start=0, stop=8)
addr
addr.value_counts()
```

```python
data_G.소재지지번주소 = addr
```

```python
data_G
```

```python
data_G_score = data_G.groupby('소재지지번주소').sum()
```

```python
data_G_score_sort = data_G_score.sort_values(by='종합점수' ,ascending=False)
```

```python
len(data_G_score_sort)
```

```python
data_G_cnt = pd.DataFrame(addr.value_counts())
```

```python
data_G_cnt.index.name = '주소'
data_G_cnt.rename(columns={'소재지지번주소': '공원'}, inplace=True)
```

```python
data_G_cnt
```

```python
data_G_cnt.to_csv('./data/경기도_공원수.csv')
```

```python
data_G = pd.merge(data_G_cnt, data_G_score_sort, left_index=True, right_index=True, how='outer')
```

```python
data_G
```

```python
data_G['최종점수'] = data_G['공원'] + data_G['종합점수']
```

```python
data_G = data_G.sort_values(by='최종점수', ascending=False)
```

```python
data_G
```

```python
data_G.rename(columns={'종합점수' : '시설점수'}, inplace=True)
```

```python
data_G
```

```python
data_G.to_csv('./data/경기도_공원점수_data.csv')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('./data/버스노선.csv', 
                  low_memory=False)
```

```python
data.head(20)
```

```python
data.groupby(['관할지역', '노선번호']).count()
```

```python
tmp = data.groupby(['관할지역','노선번호']).count()
```

```python
tmp.index[0][0]  # for문 돌리기
```

```python
li = []
for f in tmp.index :
    li.append(f[0])
          
```

```python
bus_route = pd.Series(li).value_counts()
```

```python
bus_route
```

```python
bus_route = pd.DataFrame(bus_route)
```

```python
bus_route
```

```python
bus_route.drop(index={'금천구', '강동구', '음성군', '구로구', '노원구', '봉화군' }, inplace=True)
```

```python
bus_route
```

```python
bus_route.rename(columns={0:'버스 노선수'}, inplace=True)
```

```python
bus_route
```

```python
bus_route.to_csv('./data/경기도_버스노선수_data.csv')
```

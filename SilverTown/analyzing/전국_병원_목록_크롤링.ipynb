```python
import pandas as pd
import bs4
import requests
from urllib.request import urlopen
```

```python
url = "https://www.ppm.or.kr/medi/mediInfoView.do?MENUID=A03050100"
response = requests.get(url)
```

```python
response
```

```python
response.text
```

```python
import bs4
bs_obj = bs4.BeautifulSoup(response.text, 'html.parser')
print(bs_obj.prettify())
```

```python
hospital_list = bs_obj.find_all('tr')
```

```python
hospital_list
```

```python
hospital_list[1].find_all('td')[3]
```

```python
hospital_title=[]
hospital_type=[]
hospital_addr=[]
for hospital in hospital_list:
    try:
        hospital_title.append(hospital.find_all('td')[1].a.text)
        hospital_type.append(hospital.find_all('td')[3].text)
        hospital_addr.append(hospital.find_all('td')[4].text)
    except:
        continue
    
```

```python
hospitals = {'의료기관명': hospital_title, '병원종류': hospital_type, '주소': hospital_addr}
hospital_df = pd.DataFrame(hospitals, index=range(1, len(hospital_title)+1))
hospital_df.head()
```

```python
res = hospital_df.병원종류.str.contains('종합병원', na=False)
hospital_df = hospital_df[res]
hospital_df.head(100)
```

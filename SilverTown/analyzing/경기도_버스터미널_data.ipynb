```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('./data/전국버스터미널현황.csv', 
                  low_memory=False)
```

```python
data.head(20)
```

```python
res = data.TERMN_ADDR.str.contains('경기', na=False)
data_terminal = data[res]
data_terminal
```

```python
addr = data_terminal.TERMN_ADDR
```

```python
addr = addr.str.slice(start=0, stop=8)
addr.value_counts()
```

```python
addr = addr.str.replace('경기 화성시 오', '경기도 화성시 ')
```

```python
addr = addr.str.replace('여주군', '여주시')
```

```python
addr.value_counts()
```

```python
terminal_cnt = addr.value_counts()
```

```python
terminal_cnt
```

```python
terminal_cnt = pd.DataFrame(terminal_cnt)

```

```python
terminal_cnt
```

```python
terminal_cnt.rename(columns={'TERMN_ADDR': '버스터미널'}, inplace=True)
```

```python
terminal_cnt
```

```python
terminal_cnt.to_csv('./data/경기도_버스터미널수_data.csv')
```

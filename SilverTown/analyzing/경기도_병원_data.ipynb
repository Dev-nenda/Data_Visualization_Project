```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('./data/병원list.csv', 
                  low_memory=False)
```

```python
data.head()
```

```python
res = data.주소.str.contains('경기', na=False)
data_hospital = data[res]
data_hospital.head(20)

```

```python
addr = data_hospital.주소.str.slice(start=0, stop=8)
addr

addr = addr.str.replace('경기도','경기')
addr = addr.str.replace('경기', '경기도')
addr = addr.str.replace('안산시 상', '안산시')
addr = addr.str.replace('파주시 중', '파주시')
```

```python
addr
```

```python
data_hospital.주소 = addr
```

```python
data_hospital
```

```python
data_hospital.drop(columns={'Unnamed: 0'}, inplace=True)
```

```python
data_hospital
```

```python
data_hospital.to_csv('./data/경기도_종합병원_목록.csv')
```

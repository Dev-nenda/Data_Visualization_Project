```python
import pandas as pd
import bs4
import requests
from urllib.request import urlopen
```

```python
url = "https://9goodnews.com/entry/%EC%A0%84%EA%B5%AD-%EC%8B%A4%EB%B2%84%ED%83%80%EC%9A%B4-%EB%A6%AC%EC%8A%A4%ED%8A%B8"
response = requests.get(url)
```

```python
response
```

```python
response.text
```

```python
import bs4
bs_obj = bs4.BeautifulSoup(response.text, 'html.parser')
print(bs_obj.prettify())
```

```python
silvertown_list = bs_obj.find_all('tbody')
```

```python
silvertown_list
```

```python
silvertown_list[0]
```

```python
silvertown_list[0].find_all('tr')[1].find_all('td')[1].text
```

```python
silvertown_title = []
silvertown_addr = []

for tbody in silvertown_list:
    for tr in tbody.find_all('tr')[1:] :
        silvertown_title.append(tr.find_all('td')[1].text)
        silvertown_addr.append(tr.find_all('td')[3].text)
```

```python
silvertowns = {'시설명': silvertown_title, '주소':silvertown_addr}
silvertown_df = pd.DataFrame(silvertowns, index=range(1, len(silvertown_title)+1))
```
